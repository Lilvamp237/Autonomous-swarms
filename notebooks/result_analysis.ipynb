```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load BOTH datasets
# Make sure you have both of these files in your metrics folder!
df_selfish = pd.read_csv('../data/metrics/selfish2.csv')
df_mixed = pd.read_csv('../data/metrics/altra2.csv')

# Set up the academic plot style
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# --- Plot 1: Cumulative Throughput ---
ax1.plot(df_mixed['tick'], df_mixed['throughput_vehicles'], 
         color='royalblue', linewidth=3, label='Proposed: Mixed SVO Swarm')
ax1.plot(df_selfish['tick'], df_selfish['throughput_vehicles'], 
         color='crimson', linewidth=2, linestyle='--', label='Baseline: Selfish Swarm (0°)')

ax1.set_title('Cumulative Intersection Throughput over Time', fontsize=14, fontweight='bold')
ax1.set_ylabel('Total Vehicles Arrived', fontsize=12)
ax1.set_xlabel('Simulation Step', fontsize=12)
ax1.legend(loc='upper left')

# --- Plot 2: Intersection Delay ---
ax2.plot(df_mixed['tick'], df_mixed['intersection_delay_s'], 
         color='seagreen', linewidth=2, label='Mixed SVO Swarm')
ax2.plot(df_selfish['tick'], df_selfish['intersection_delay_s'], 
         color='crimson', linewidth=2, linestyle='--', alpha=0.8, label='Selfish Gridlock')

ax2.set_title('Aggregate Intersection Delay', fontsize=14, fontweight='bold')
ax2.set_ylabel('Total Waiting Time (seconds)', fontsize=12)
ax2.set_xlabel('Simulation Step', fontsize=12)
ax2.legend(loc='upper left')

# --- Plot 3: Swarm Mean Speed ---
ax3.plot(df_mixed['tick'], df_mixed['mean_speed_ms'], 
         color='darkorange', linewidth=2, label='Mixed SVO Swarm')
ax3.plot(df_selfish['tick'], df_selfish['mean_speed_ms'], 
         color='crimson', linewidth=2, linestyle='--', alpha=0.8, label='Selfish Swarm')

ax3.set_title('Swarm Mean Velocity', fontsize=14, fontweight='bold')
ax3.set_ylabel('Speed (m/s)', fontsize=12)
ax3.set_xlabel('Simulation Step', fontsize=12)
ax3.legend(loc='upper right')

# Format and save
plt.tight_layout()
plt.savefig('comparative_swarm_results.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your finalized datasets (Make sure names match your files!)
df_selfish = pd.read_csv('../data/metrics/selfish5.csv') 
df_mixed = pd.read_csv('../data/metrics/mixed5.csv')

plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 9))

# =======================================================
# PANEL 1: VELOCITY COLLAPSE (The Traffic Jam Highlight)
# =======================================================
ax1.plot(df_selfish['tick'], df_selfish['mean_speed_ms'], color='crimson', linewidth=2, label='Baseline: Selfish Swarm')
ax1.plot(df_mixed['tick'], df_mixed['mean_speed_ms'], color='royalblue', linewidth=2, label='Proposed: Mixed SVO Swarm')

# MAGIC TRICK: Shade the background RED whenever the selfish swarm drops below 1.5 m/s (Gridlock Phase)
ax1.fill_between(df_selfish['tick'], 0, df_selfish['mean_speed_ms'], 
                 where=(df_selfish['mean_speed_ms'] < 1.5), 
                 color='red', alpha=0.3, label='Critical Traffic Jam / Deadlock')

ax1.set_title('Swarm Velocity Collapse (Traffic Jams)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Mean Speed (m/s)', fontsize=12)
ax1.legend(loc='upper right')

# =======================================================
# PANEL 2: DELAY MOUNTAINS & COLLISIONS
# =======================================================
# Fill the area to show the massive weight of the delay
ax2.fill_between(df_selfish['tick'], 0, df_selfish['intersection_delay_s'], color='crimson', alpha=0.5, label='Selfish Delay Accumulation')
ax2.fill_between(df_mixed['tick'], 0, df_mixed['intersection_delay_s'], color='royalblue', alpha=0.7, label='Mixed SVO Delay')

# Explicitly plot Collisions as big "X" markers if they occurred
selfish_crashes = df_selfish[df_selfish['collisions'] > 0]
if not selfish_crashes.empty:
    ax2.scatter(selfish_crashes['tick'], selfish_crashes['intersection_delay_s'], 
                color='black', marker='X', s=150, zorder=5, label=f'Collisions ({len(selfish_crashes)} events)')

ax2.set_title('System Delay & Collision Events', fontsize=14, fontweight='bold')
ax2.set_ylabel('Total Waiting Time (s)', fontsize=12)
ax2.set_xlabel('Simulation Step', fontsize=12)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.savefig('gridlock_collision_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
